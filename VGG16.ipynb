## using keras

```python
from keras.applications import VGG16
from keras.applications.vgg16 import preprocess_input, decode_predictions

from keras.preprocessing.image import load_img, img_to_array
from tensorflow import expand_dims

import matplotlib.pyplot as plt
```

```python
keras_model = VGG16(weights='imagenet')
```

```python
img = load_img("/content/tinca tinca.jpg", target_size=(224, 224, 3))
plt.imshow(img)
```

```python
arr_img = img_to_array(img)
arr_img.shape
```

```python
arr_img = expand_dims(arr_img, axis=0)
arr_img.shape
```

```python
vgg_img = preprocess_input(arr_img)
y_pred = keras_model.predict(vgg_img)
print(y_pred)
```

```python
predictions = decode_predictions(y_pred, top=5)
predictions
```

```python
path_image = ["tinca tinca.jpg", "white shark.jpg", "african lion.jpg"]
for img in path_image:
  image = load_img(img, target_size=(224, 224, 3))
  plt.imshow(image)
  plt.axis('off')
  plt.pause(1)
  plt.clf()
  arr_img = img_to_array(image)
  arr_img = expand_dims(arr_img, axis=0)
  vgg_img = preprocess_input(arr_img)
  y_pred = keras_model.predict(vgg_img)
  predictions = decode_predictions(y_pred)
  print(predictions)
```

## using pytorch

```python
import torch
from torchvision.models import vgg16
from torchvision import transforms
from PIL import Image
```

```python
pytorch_model = vgg16(pretrained=True)
```

```python
#batch_normalization & Drop_out freezing
pytorch_model.eval()

#input preprocess
preprocess_transform = transforms.Compose([transforms.Resize(256),
                                           transforms.CenterCrop(224),
                                           transforms.ToTensor(),
                                           transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                                                std=[0.229, 0.224, 0.225])
                                           ])

img = Image.open("/content/african lion.jpg")
proc_img = preprocess_transform(img)
proc_img.shape
```

```python
plt.imshow(proc_img.cpu().permute(1, 2, 0).numpy())
```

```python
torch_img = torch.unsqueeze(proc_img, axis=0)
torch_img.shape
```

```python
with torch.no_grad():
  pred = pytorch_model(torch_img)

pred
```

```python
_, indices = torch.sort(pred, descending=True)
print(indices[0][0:5])
```

```python
import json
from urllib import request

url = "https://storage.googleapis.com/download.tensorflow.org/data/imagenet_class_index.json"
imagenet_classes = json.loads(request.urlopen(url).read().decode())

for idx in indices[0][0:5]:
  class_id = str(idx.item())
  print(imagenet_classes[class_id][1])
```

```python
path_image = ["tinca tinca.jpg", "white shark.jpg", "african lion.jpg"]
for img in path_image:
  image = Image.open(img)
  proc_img = preprocess_transform(image)
  plt.imshow(proc_img.cpu().permute(1, 2, 0).numpy())
  plt.axis('off')
  plt.pause(1)
  plt.clf()
  torch_img = torch.unsqueeze(proc_img, 0)
  with torch.no_grad():
    pred = pytorch_model(torch_img)
  _, indices = torch.sort(pred, descending=True)
  for idx in indices[0][0:5]:
    class_id = str(idx.item())
    print(imagenet_classes[class_id][1])
```
